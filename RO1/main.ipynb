```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB, GaussianNB, MultinomialNB
from sklearn.svm import SVC
import pandas as pd
import numpy as np


df = pd.read_csv('data/mammographic_masses_data.csv', delimiter=',', names=['in_1', 'in_2', 'in_3', 'in_4', 'in_5', 'out_1'])


display(df)

```

```python
df = df.replace('?', np.NaN)
if not df.isnull().values.any():
    display("This DataFrame doesn't exist null values")
else: 
    df = df.dropna(how='any',axis=0)
    display('Rows with null values was deleted from the DataFrame')

X = df.copy().drop(columns='out_1').to_numpy()
y = df['out_1'].copy().to_numpy()
test_sizes_list = [0.4, 0.3, 0.2, 0.1]

result_df = pd.DataFrame(columns = ['Classifier', 'Test size', 'Accuracy'])

for clf in BernoulliNB(), GaussianNB(), GradientBoostingClassifier(), SVC(), MultinomialNB():
    
    for ratio in test_sizes_list:

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=ratio)
        clf.fit(X=X_train, y=y_train)
        prediction = clf.predict(X_test)

        new_row = {'Classifier':str(clf.__class__.__name__), 'Test size':ratio,'Accuracy':clf.score(X_test, y_test)}

        result_df = result_df.append(new_row, ignore_index = True)

        
display(result_df)
result_df.to_excel("RO_LAB1_RESULTS.xlsx", index=False, sheet_name='Data', engine='xlsxwriter')
```
